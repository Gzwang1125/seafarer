```python
from seafarer.models import Histogram, BasePlot
from seafarer.colors import ihme_palettes
from seafarer.utils import sort, make_plot, save_plot, edit_labels
import seaborn as sns 

testpal = ['#D32B47', "#22973D"] * 3

tips = sns.load_dataset('tips')

tips.head()
```

```python
df = tips.copy(deep=True)
df = sort(df, 'day')

plot = make_plot("histogram", df=df, x='day', y="day", 
          dropna=True, width=5, height=5)
plot.title("A crappy title")
plot.xticklabels(df, 'day', rotation=45)
edit_labels(xlabel="Day of the week!", ylabel="Count~~")
save_plot(file_path='test.svg')
```

```python
plot = make_plot("swarm_boxplot", df=df, y='day', x="tip", 
          dropna=True, custom_palette="Set2")
plot.title('A nice title')
```

```python
plot = make_plot("violin-plot", df=df, y='day', x="tip", height=4, width=4,
          dropna=True, custom_palette='1')
plot.title('A nice title')
```

```python
plot = make_plot("scatterplot", df=df, y='total_bill', x="tip", height=4, width=4,
          hue='sex', custom_palette='33', dropna=True)
```
